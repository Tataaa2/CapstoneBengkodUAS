```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
display(df.head())
```

```python
print(df.info())
```

```python
display(df.describe())
```

```python
missing = df.isnull().sum()
print(missing)

missing.plot(kind='bar', figsize=(10, 5))
plt.title("Missing Value per Kolom")
plt.show()
```

```python
df['Churn'].value_counts().plot(kind='bar', figsize=(5,4))
plt.title("Distribusi Churn")
plt.show()
```

```python
print("\nDistribusi Numerik:")
display(df.describe())
```

```python
df_num = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8,6))
sns.heatmap(df_num.corr(), annot=True, cmap='Blues')
plt.title("Heatmap Korelasi Numerik")
plt.show()
```
